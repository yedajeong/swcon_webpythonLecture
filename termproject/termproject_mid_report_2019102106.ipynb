```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = ("https://www.metacritic.com/browse/games/genre/name/action/pc?view=condensed")
req = requests.get(URL)
print(req.status_code)
```

```python
c=req.content
```

```python
soup = BeautifulSoup(c,"html.parser")
print(soup)
```

```python
gameName = soup.select('div[class=tab_item_name]')
print(gameName)
```

```python
for tag in gameName:
    print (tag.text)
```

```python
def url_call(userUrl):
    req = requests.get(userUrl)
    userContent=req.content
    soup = BeautifulSoup(userContent,"html.parser")
    gameName = soup.select('span[class=title]')
    for tag in gameName:
        print (tag.text)
```

```python
for i in range(2,31):
    URL = 'https://store.steampowered.com/search/?filter=topsellers&page='+str(i)+'&tags=19&category1=998'
    print(url_call(URL))
```
