# ``matplotlib`` 라이브러리 이해

``matplotlib``은 고품질의 수치 기반 그래프를 생성하는 Python 라이브러리입니다.<br>
``matplotlib``는 Jupyter 노트북 등 여러 그래픽 사용자 인터페이스 툴킷에서 사용할 수 있습니다.<br>
``matplotlib``은 단지 몇 줄의 코드로 플롯, 히스토그램, 파워 스펙트럼, 막대 차트, 오류 차트, 산점도 등을 생성 할 수 있습니다. <br>
간단한 그래프를 나타내기 위해 ``matplotlib``안의 ``pyplot`` 모듈은 특히 IPython과 결합 할 때 MATLAB과 같은 인터페이스를 제공합니다. <br>
고급 사용자의 경우 객체 지향 인터페이스 또는 MATLAB 사용자에게 익숙한 함수 세트를 통해 선 스타일, 글꼴 속성, 축 속성 등을 완벽하게 제어 할 수 있습니다.<br>

본 수업에서는 ``matplotlib``이 할수 있는 능력의 맛보기만 보여줍니다.<br>
``matplotlib``의 공식 웹 사이트( https://matplotlib.org/index.html )를 방문하면, 더 많은 정보와 예제를 볼 수 있습니다.<br>

아래에 사용방법을 직관적으로 이해할 수 있는 간단한 예제들을 만들어서 실행할 수 있도록 하였습니다.<br>
프로그램을 읽고 어떤 의미일지 추측하고, 아울러 공식 사이트 등에서 의미를 정확하게 파악도 하고, 실제 코드를 조금씩 수정하면서 이해해 봅니다.

## Line Chart 만들기

```python
%matplotlib inline
from matplotlib import pyplot as plt

X_axis = ['1st','2nd', '3rd', '4th', '5th', '6th']
Y_axis = [10, 20, 40, 80, 160, 320]

plt.plot(X_axis, Y_axis, color='red', marker="o", linestyle='solid')

plt.title("Line Chart")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Multi Line Chart 만들기

```python
%matplotlib inline
from matplotlib import pyplot as plt

X_axis = ['1st','2nd', '3rd', '4th', '5th', '6th']
Y1_axis = [10, 20, 40, 80, 160, 320]
Y2_axis = [320, 160, 80, 40, 20, 10]
Y3_axis = [10, 60, 110, 160, 210, 260]

xs = [i for i, _ in enumerate(Y_axis)]

plt.plot(xs, Y1_axis, 'r-', label="Y1_axis")
plt.plot(xs, Y2_axis, 'g-', label="Y2_axis")
plt.plot(xs, Y3_axis, 'b:', label="Y3_axis")

plt.legend(loc=9)

plt.title("Multi Line Chart")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Bar Chart (Histogram) 만들기

```python
%matplotlib inline
from matplotlib import pyplot as plt

X_axis = ['1st','2nd', '3rd', '4th', '5th', '6th']
Y_axis = [10, 20, 40, 80, 160, 320]

xs = [i for i, _ in enumerate(Y_axis)]

plt.xticks([i for i, _ in enumerate(X_axis)], X_axis)

plt.bar(xs,Y_axis)

plt.title("Bar Chart (Histogram)")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Scatter Chart 만들기

```python
%matplotlib inline
from matplotlib import pyplot as plt

X_axis = ['1st','2nd', '3rd', '4th', '5th', '6th']
Y_axis = [10, 20, 40, 80, 160, 320]

labels = [10, 20, 40, 80, 160, 320]

plt.scatter(X_axis, Y_axis)

for label, X_axis_count, Y_axis_count in zip(labels, X_axis, Y_axis):
    plt.annotate(label,
                xy=(X_axis_count, Y_axis_count),
                xytext=(5,-5),
                textcoords='offset points')

plt.title("Scatter Charts")
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# ``matplotlib`` 공식 사이트의 일부 예제 구경하기

앞서 명시한 ``matplotlib`` 공식 사이트에는 다양한 활용 방법에 대한 예제들이 있습니다.<br>
별도의 고가 소프트웨어를 사용하지 않아도 Python 기반의 ``matplotlib`` 라이브러리에서 충부한 그래프 작성 기능을 제공하므로,<br>
이에 대한 사전 지식을 가지고 있다면, 본인이 확보한 데이터의 의미를 명확하게 이해하고 잘 활용할 수 있을 겁니다.<br>

## Simple Plot 예제

```python
# Source: https://matplotlib.org/gallery/lines_bars_and_markers/simple_plot.html

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```

## 3D Surface 예제

```python
# Source: https://matplotlib.org/gallery/mplot3d/surface3d.html

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Pie Chart 예제

```python
# Source: https://matplotlib.org/gallery/pie_and_polar_charts/pie_features.html

%matplotlib inline
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Log Chart 예제

```python
# Source: https://matplotlib.org/gallery/scales/log_demo.html

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
t = np.arange(0.01, 20.0, 0.01)

# Create figure
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

# log y axis
ax1.semilogy(t, np.exp(-t / 5.0))
ax1.set(title='semilogy')
ax1.grid()

# log x axis
ax2.semilogx(t, np.sin(2 * np.pi * t))
ax2.set(title='semilogx')
ax2.grid()

# log x and y axis
ax3.loglog(t, 20 * np.exp(-t / 10.0), basex=2)
ax3.set(title='loglog base 2 on x')
ax3.grid()

# With errorbars: clip non-positive values
# Use new data for plotting
x = 10.0**np.linspace(0.0, 2.0, 20)
y = x**2.0

ax4.set_xscale("log", nonposx='clip')
ax4.set_yscale("log", nonposy='clip')
ax4.set(title='Errorbars go negative')
ax4.errorbar(x, y, xerr=0.1 * x, yerr=5.0 + 0.75 * y)
# ylim must be set after errorbar to allow errorbar to autoscale limits
ax4.set_ylim(ymin=0.1)

fig.tight_layout()
plt.show()
```
